```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

# Email credentials
email_address = 'jadhavarjun376@gmail.com'
password = 'Arjun@123'

# Email template
subject = "Automation Test Engineer - Cypress Expertise"
body = """
Dear [Hiring Manager],

I am excited to submit my application for the Automation Test Engineer role requiring Cypress skills. With over 2.5+ years of hands-on test automation experience using the Cypress framework, I believe I would excel in this position.

As my resume shows, I have:

• Developed end-to-end test automation suites for two web applications using Cypress and Mocha framework, enhancing test coverage
• Created and managed test cases using JIRA and reported defects efficiently
• Expert-level JavaScript knowledge to write custom commands, mocks, stubs for better testability
• Strong expertise in writing and implementing Page Object Model design pattern
• Good working knowledge of CI/CD pipelines, having integrated automated suites into tools like Jenkins

I also have exposure to:
- Creating automated API tests using cy.request()
- BDD-style test framework development with Cucumber
- Cross-browser testing strategies

With my in-depth Cypress skills and passion for creating robust, browser-based test automation, I am well-aligned to the job responsibilities. I would welcome the chance to bring my technical expertise in automated testing along with adaptability and strong work ethic to your dynamic company.

I have attached my latest resume and look forward to further discussions. Please let me know if you need any other details. Thank you for your consideration.

Best regards,
Arjun Jadhav
(Automation Test Engineer)
"""

# Attachment file path
resume_file = "C:/Users/hp/Desktop/"  

# Function to send email
def send_email(to_email):
    msg = MIMEMultipart()
    msg['From'] = email_address
    msg['To'] = to_email
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'plain'))

    # Attach resume
    attachment = open(resume_file, "rb")
    part = MIMEBase('application', 'octet-stream')
    part.set_payload((attachment).read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', "attachment; filename= " + resume_file)
    msg.attach(part)

    # Connect to Gmail's SMTP server
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(email_address, password)

    # Send email
    text = msg.as_string()
    server.sendmail(email_address, to_email, text)

    # Close the server connection
    server.quit()

# List of HR Manager email addresses
hr_managers = ["aj765500@gmail.com"]

# Sending emails
for hr_manager_email in hr_managers:
    send_email(hr_manager_email)
    print(f"Email sent to {hr_manager_email}")

print("All emails sent successfully.")
```
